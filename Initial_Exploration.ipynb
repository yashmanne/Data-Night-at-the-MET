```python
clean_data = lib_data.copy(deep=True)
```

```python
clean_data['Gallery Number'].replace("in Great Hall", 1001, inplace=True)
clean_data['Gallery Number'].replace("Petrie Ct. Café", 1002, inplace=True)
clean_data['Gallery Number'].replace("on Fifth Avenue", 1003, inplace=True)
clean_data['Gallery Number'].replace("Watson Library", 1004, inplace=True)
clean_data['Gallery Number'] = clean_data['Gallery Number'].astype(float)
clean_data['Gallery Number'] = clean_data['Gallery Number'].astype("Int64")

clean_data['AccessionYear'].replace(19171917.0, 1917, inplace=True)
clean_data['AccessionYear'].replace("2005-02-15", 2005, inplace=True)
clean_data['AccessionYear'].replace("2020-03-23", 2020, inplace=True)
clean_data['AccessionYear'].replace("1953-05-28", 1953, inplace=True)
clean_data['AccessionYear'].replace("1956-08-24", 1956, inplace=True)
clean_data['AccessionYear'].replace("1926-06-07", 1926, inplace=True)
clean_data['AccessionYear'] = clean_data['AccessionYear'].astype(float)
clean_data['AccessionYear'] = clean_data['AccessionYear'].astype("Int64")
```
